```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv(r'D:\comicstaan\movies.csv')
```

```python
df.head()
```

```python
#find missing values
for col in df.columns:
     missing=df[col].isnull().sum()
     print('{}-{}'.format(col,missing))   
```

```python
#drop missing values
df=df.dropna()
```

```python
#recheck for missing values
for col in df.columns:
     missing=df[col].isnull().sum()
     print('{}-{}'.format(col,missing))   
```

```python
df.dtypes
```

```python
df['votes']=df['votes'].astype('int64')
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
df['runtime']=df['runtime'].astype('int64')
```

```python
df.head()
```

```python
#to make released and year match
df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

```python
df.head()
```

```python
df=df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#display complete data
pd.set_option('display.max_rows',None)
```

```python
df[df.duplicated()]
```

```python
#scatter plot with budget and gross
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('budget vs gross')
plt.xlabel('budget')
plt.ylabel('gross')
plt.show()
```

```python
df.head(20)
```

```python
#plot regression plot of budge vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'yellow'})
```

```python
#lets start looking at correlation
df.corr()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix for numerical features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#lets correlate company and gross
#so change the obejects(names) to numbers
df_numbers=df
for col_name in df_numbers.columns:
    if(df_numbers[col_name].dtype == 'object'):
        df_numbers[col_name]=df_numbers[col_name].astype('category')
        df_numbers[col_name]=df_numbers[col_name].cat.codes 
df_numbers=df.sort_values(by=['gross'], inplace=False, ascending=False)
df_numbers.head(30)
```

```python
correlation_matrix = df_numbers.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix for numerical features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numbers.corr()
```

```python
correlation_pairs=df_numbers.corr().unstack()
correlation_pairs
```

```python
sorted_pairs=correlation_pairs.sort_values()
sorted_pairs
```

```python
high_corr= sorted_pairs[sorted_pairs > 0.5]
high_corr
```
